### Cloud SQL<br>

***

<br>

**Cloud SQL**<br>

- a fully managed, cloud native relational database service (RDBMS)<br>
that offers MySQL, postgres, SQL server engine with builtin support for replication<br>
- DBaaS (Database as a Service)<br>
- Low latency, transactional, relational DB workloads<br>
- MySQL, PostgreSQL, SQL Server<br>
- Offers replication - read replicas<br>
- High availability<br>
- On-demand and auto backups<br>
- Point in time recovery - restore DB in specific point in time<br>
- up to 30 TB storage capacity<br>
- Automatic storage increase<br>
- encryption at rest and in transit<br>
when it comes to cost, you're billed for CPU, memory, storage, egress traffic<br>

Machine type: (Shared-core, Standard, High Memory)<br>
Storage type: HDD or SSD (up to 30TB)<br>
<br>

**Connect to Cloud SQL:**<br> 
- public or private IP, private IP is preferred<br>
- Cloud SQL Proxy allows you to authorize a connection using IAM permissions<br>
- Unless the connection comes from an authorized network: the user specified it<br>
- External Application, like: Compute Engine, GKE, App Engine, Cloud Functions, Cloud Run<br>

(Client App <-local connection-> Proxy Client) <--TCP secure tunnel--> (Proxy Server <-> Instance(with TCP standard port))<br>
<br>

**Cloud SQL Proxy**<br>
uses IAM permissions to authorize connections<br>
proxy validates connection using credentials for a user or service account<br>
wrapping the connection in SSL/TLS layer, authorized for Cloud SQL instance<br>
using Cloud SQL Proxy is the recommended method for connecting connections for Cloud SQL<br>
provides a secure access and uses TCP connection method<br>

it's open source and can be executed as a binary, available in Linux, MacOS, Windows<br>
Cloud Proxy act as an intermediatory server<br>
That listens to incoming connections wraps, them in SSL/TLS then passes them to Cloud SQL instance<br>
Cloud Proxy handles the authentication with Cloud SQL, providing secure access w/o managing IPs<br>
<br>

**Replication**<br>
replication scales the use of data in database w/o degrading performance<br>
migrating data between regions, platforms or on-premises database to Cloud SQL<br>
you can promote a replica if the original instance become corrupted<br>

US-EAST1-B - Cloud SQL (Primary Instance) -> US-EAST1-C - Read Replica<br>
Read replicas are read-only, can't write to them<br>
read replicas processes queries, read requests, and analytics traffics, reducing the load on primary instance<br>
read replicas can have more CPUs and memory than the primary instance, but they can't have any less<br>
you can have up to 10 read replicas per primary instance<br>
you can connect to its replica directly using connection name and IP<br>
<br>

**Read Replica types:**<br>
Cloud SQL supports: read replicas, cross-region read replicas, external read replicas<br>
- **read replica** - used to offload work from Cloud SQL instance<br>
exact copy of primary instance<br>
- **cross region read replica** - bring closer to your region<br>
and disaster recovery capabilities and migration of data with minimum downtime<br>
- **external read replica** - external mySQL instances that replicate from primary instance<br>

**to create read replicas:**<br>
- automated backups must be enabled<br>
- binary logging must be enabled<br>
- point in time recovery has to be enabled<br>
- one backup must have been created after binary logging was enabled<br>
<br>

**Promoting Replicas**<br>
you can promote a read replica / cross region read replica<br>
converts to primary instance with read/write capability<br>

this can't be undone and other replicas are NOT transferred over from primary instance<br>
you'd have to reconnect to the new replica<br>

**reason for promoting replicas**<br>
- regional migration (planned)<br>
- disaster recovery (unplanned)<br>
<br>

**High availability**<br>

Primary Instance - Standby Instance<br>
through synchronous replication, all write made to primary instance is made to standby instance<br>

if primary instance becomes unavailable, Failover is initiated<br>
and switches over to standby instance<br>
It'll failback when it comes back<br>

HH configured instance charged at double the price than standalone instance<br>
standby instance can't be used for read-queries<br>
automatic back up and point-in-time recovery must be enabled for high availability<br>
<br>

**Back ups**<br>

By default, Cloud stores back up data in 2 region, 1 is in the region SQL is in, the other in another region<br>

**Types of backups:**<br>
**on-demand** - create backup at any time, persist until you delete<br>
**automated** - 4 hour backup window, occur everyday, 7 most recent backups are retained<br>
should schedule to back up when it's not busy<br>
**point in time recovery(PITR)** - recover an instance to a specific point-in-time, always creates a new instance<br>
