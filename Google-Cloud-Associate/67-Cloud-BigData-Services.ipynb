### Cloud Big Data Services<br>

***

<br>

**What's Big Data?**<br>

- Massive amounts of data<br>
- characteristics of Big Data on traditional databases<br>
    - too expensive to store<br>
    - traditional databases are not cost effective<br>
    - no flexibility for storing unstructured data<br>
    - cannot accommodate real time data<br>
    - lack support for Petabytes-scale of data volumes<br>

Apache Hadoop and NooSQL to rescue<br>
but it's extremely complex to deploy and manage<br>
<br>

**Why is Big Data so important?**<br>

help companies make better decisions - business value<br>
- gain useful insights<br>
- increase revenues<br>
- get/retain customers<br>
- improve operations<br>
- better with machine learning<br>
<br>

**Big Data Services**<br>

**BigQuery**<br>
BigQuery is a fully managed, serverless database, with petabyte scale, low cost analytics data warehouse<br>
Bigdata would ingest that data, by doing a batch uploads and streaming it in real time<br>
you can do manual ingestion or pub/sub ETL data<br>
data transfer service, you can automatically transfer data from external google data sources, partner SaaS apps to bigQuery<br>
batch and exports are free services<br>

- real time analytics<br>
- automatic high availability<br>
- automatic backups and restore<br>
- standard SQL<br>
- big data ecosystem integration<br>
- web UI, bq, API -> data studio sheets looker<br>
- data governance<br>
- geo-expansion<br>
- IAM, VPC management<br>
- Data encryption<br>

on-demand pricing charged for bytes read, not bytes returned<br>
<br>

**Pub/Sub**<br>
Pub/Sub is a fully managed, real-time messaging service, send receive messages between independent applications<br>
publisher -sends a message-> (Pub/Sub) Topic -> Message Storage<br>
Topic -> Subscription -> message -> subscriber -> sends acknowledgement<br>
<br>

**Composer**<br>
Composer is a fully managed workflow orchestration service, built on Apache Airflow<br>
<br>

**Dataflow**<br>
Dataflow is a serverless, fully managed processing service for executing Apache Beam pipelines for batch and realtime data streaming<br>

dataflow job: ETL (Export, Transport, Load)<br>

Realtime events/IoT -> Cloud Pub/Sub -streaming pipeline-> Cloud Dataflow (parallel data processing)<br>
-streeaming pipeline-> bigQuery (analytics engine) -> Cloud Datalab/Spreadsheet/Co-workers<br>

Database -> Cloud Storage -batch pipeline-> Cloud Dataflow -batch pipeline-> BigQuery<br>
-> Cloud Datalab/Spreadsheet/Co-workers<br>
<br>

**Dataproc**<br>
Dataproc is a fully managed Spark and Hadoop service<br>
Can be used to replace on-prem Hadoop infrastructure<br>
- Fast<br>
- Integrated<br>
- Managed<br>
- Same tools<br>
- 1 cent/vCPU<br>
<br>

**Dataproc vs. Dataflow**<br>

Dataproc<br>
- Managed<br>
- Dependencies to tools in the Hadoop Ecosystem<br>

Dataflow<br>
- Serverless<br>
- Apache Beam Runtime - more portable<br>
<br>

**Cloud Datalab**<br>
Interactive developer tool created to explore analysis, explore and visualize data<br>
and build ML model from your data, uses open-sourced Jupyter notebook<br>
It runs on Compute Engine and connects to multiple Cloud Services easily<br>
<br>

**Cloud Dataprep**<br>
Serverless intelligent data service for visually exploring, cleaning and preparing structured<br>
and unstructured data for analysis, reporting and machine learning<br>
<br>
