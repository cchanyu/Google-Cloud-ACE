### Cloud App Engine Overview<br>

***

<br>

**App Engine** is a fully managed, serverless platform to develop/host web applications<br>
PaaS Service, designed for developers, App Engines do heavy liftings with provisioning and scaling<br>
Code or containers - Python, Java, Node.js, Go, Ruby, PHP, or .NET<br>
Autoscaling based on load<br>
Versions - allow rollbacks, migrating or traffic splitting<br>
Supports connecting to external storage, such as: Cloud Firestore, Cloud SQL, Cloud Storage<br>
and on-Premises Databases or even external databases on other Clouds<br>
Standard and flexible environments<br>
<br>

**Standard and Flexible Environments**<br>

App Engine - Standrard<br>
- Apps run in sandbox environment<br>
- Specific versions of runtimes used<br>
- Run for free or at very low cost<br>
- designed for sudden and extreme spikes of traffics<br>
- pricing based on instance hours<br>

App Engine - Flexible<br>
- app runs in docker containers<br>
- any versions of runtimes used<br>
- no free quota available<br>
- designed for consistent traffics<br>
- pricing based on VM resources<br>
- Managed VMs<br>
<br>

**Deploying an Application**<br>

Application (Top level) -> Services 1/2 -> Versions 1/2 -> Instance 1/2<br>
> gcloud app deploy<br>

auto builds a container image from config file<br>
by using Cloud Build service and deploys that image to App Engine<br>

Each service can be configured to use different runtimes<br>
and to operate with different performance settings<br>

services in App Engine are used to factor large applications<br>
into logical components that can securely share App Engine features<br>
and communicate with one or another<br>
these app engine services become loosely-coupled behaving like micro-services<br>

within each services, you deploy versions<br>
each versions then runs within one or more instances<br>
depending on how much traffic you configured it to handle<br>

having multiple versions of your application within each service<br> 
allows you to quickly switch between different versions of that application<br>
for rollbacks, testing or other temporary events<br>

you can route traffics one or more specific versions of your applications<br>
by migrating traffic to one specific version or splitting your traffic between 2 versions<br>

versions within your services run 1 or more instances<br>
by default app engine scale your app to match the load<br>
<br>

**Managing Instances**<br>

auto create and shut down instances as traffic fluxate<br>
or you can specify the number of instances to run<br>
specify scaling type for your app in app.yaml<br>
- automatic scaling - based on metrics like request rate and response latencies<br>
- basic scaling - creates instances when your app receives requests<br>
- manual scaling - specifies the number of instances continuously run regardless of the load<br>
<br>

**Traffic Migration**<br>

Traffic migration switches the request routing between the versions within a service of your app<br>
moving traffic from one or more versions to a single new version.<br>
when deploying, new version with the same name, causes an immediate traffic migration<br>
without any running instances, your application could have a spike in latencies.<br>
<br>

**Traffic Splitting**<br>

you can decide how you want to distribute the traffic<br>
ver2 10%, ver1 90% - allows you to conduct ab testing<br>
you must choose to split by either IP address or HTTP cookie or randomly<br>
