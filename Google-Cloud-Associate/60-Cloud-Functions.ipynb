### Cloud Functions<br>

***

<br>

##### **Cloud Functions**<br>

**Serverless** execution environment, like App Engine<br>
meaning no need to provision any server, or updating VMs, as the infrastructure is all handled by Google<br>
unlike App Engine, you'll never see the server<br>

**FaaS** - Function as a Service<br>
this is where you upload code purposely in a supported language<br>
when your code is trigger, it's executed in a fully managed environment<br>

Cloud Functions runs in a **runtime** environment - Python, Java, Node.js, Go, .NET core<br>

**Event-driven** - so when something happens in your environment, you can choose whether to respond to the event<br>
Triggers - HTTP, Pub/Sub, Cloud Storage, Firestore, Firebase<br>

**Billing** - time + resources provisioned (memory)<br>

There's **free tier**, which allows 2 million invocations of your function<br>
<br>

**How Cloud Functions work**<br>

Trigger -> Written Code (+ optional settings) -> HTTP/Background functions -> bind the trigger -> Cloud Build<br>

After selecting name and region that you want the function to live in<br>
You'd then select the trigger you wish to use<br>
Ex. Triggers - HTTP, Pub/Sub, Cloud Storage, Firestore, Firebase<br>

Trigger is a declaration that you're interested in a certain event<br>
binding a function to the trigger allows you to capture and act on these events<br>

authentication configuration is the next step<br>
and can be selected with a public access<br>
or configured through IAM<br>

there are optional settings that can be configured<br>
you'd provide the amount of memory the function will need to run<br>
networking preferences<br>

once the selections been solidified,<br>
written code can be put into the function<br>
it supports variety of languages: Python, Java, Node.js, Go, .NET core<br>

There's 2 distinctive type of Cloud Functions that you could use:<br>
**HTTP functions and Background functions**<br>

with HTTP functions, you invoke them from standard HTTP requests<br>
these requests wait for the response and support handling of common HTTP request methods<br>
like GET, PUT, POST, DELETE, OPTIONS<br>
TLS certificate is automatically provisioned for you<br>
so all HTTP functions can be invoked via secure connection<br>

when it comes to background connections<br>
these are used to handle events from GCP infrastructure<br>
such as messages on Pub/Sub topic, or changes in a Cloud Storage bucket<br>
<br>

Now you're ready to deploy your code<br>
first one is binding of trigger to your function<br>
once you bind the trigger, you can't bind another one to the same function<br>
only 1 trigger can be bound at a time<br>

second thing that'll happen<br>
source code will be stored in a Cloud Storage bucket as a zip file<br>
Cloud Build, auto builds your code into a container image<br>
pushes that image to the container registry<br>
Cloud function accesses this image, when it needs to run the container to execute your function<br>
<br>

After the build is complete, we now wait for an event to happen<br>
events are things that happen within your cloud environment that you might want to take an action on<br>
these are like: data changes in Cloud SQL, files added to Cloud Storage, or new VM being created<br>
and other services like: Bigquery, Cloud SQL, Cloud Spanner<br>

when an event triggers the execution of your Cloud function<br>
data associated with the event is passed via function's parameters<br>
the type of the event determines the parameters that are passed to your function<br>

Cloud function handles incoming requests by assigning them to instances of your function<br>
Depending on the volume of the request, as well as number of existing function instances<br>
Cloud function may assign request to an existing instance or create a new one<br>

Cloud function will grab the image from Cloud registry and hand off the image along with the event data<br>
to instances for processing<br>

Each instances of function will handle only 1 concurrent request at a time<br>

fuctions has to be stateless, they aren't meant to be persistent nor is the data passed on to the function<br>
after function has run, and processed, it's passed on to VPC or to the internet<br>
by default, functions have public internet access, unless configured otherwise<br>
functions can be private and used within VPC, but must be configured before the deployment<br>
<br>
