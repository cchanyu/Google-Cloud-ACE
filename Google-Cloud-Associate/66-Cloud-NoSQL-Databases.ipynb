### Cloud NoSQL Databases<br>

***

<br>

There are 4 managed NoSQL databases in Google Cloud<br>
- **Cloud BigTable**<br>
    - fully managed, wide column noSQL database, designed for TB to PB scale workloads that offers low latency and high throughput<br>
    - built for real-time app serving and large scale analytical workloads<br>
    - regional service<br>
    - automated replication<br>
    - store large amounts of single-keyed data<br>
    - add nodes when you need them<br>
    - cluster resizing<br>
    - ideal data source for MapReduce operation: Hadoop, Apache Database<br>
    - High-priced: $450 for a node with no data<br>
    - Use cases:<br>
        - time series data<br>
        - marketing data<br>
        - financial data<br>
        - IoT data<br>
        - Graph data<br>
    - Storage Engine uses:<br>
        - Batch mapReduce operations<br>
        - Stream processing/analytics<br>
        - Machine-learning applications<br>
<br>

- **Cloud Datastore**<br>
    - fully managed, highly scalable noSQL document database built for automatic scaling, high performance, easy of app dev<br>
    - high-availability of read/writes<br>
    - atomic transactions<br>
    - automatic scaling<br>
    - SQL-like query language (GQL)<br>
    - Strong and eventual consistency<br>
    - data auto encryption at rest<br>
    - **Being retired in favour of Cloud Firestore in 2021**<br>
    - datastore emulator<br>
        - provides local emulation of the production datastore environment<br>
        - components of the Google Cloud SDKs gcloud tool<br>
    - Use cases:<br>
        - product catalogs<br>
        - user profiles<br>
        - transactions based on ACID properties, ex. transferring funds from a bank acct to another<br>
<br>

- **Firestore for Firebase**<br>
    - flexible, scalable NoSQL cloud database to store and sync data for client and server-side development<br>
    - available for Native, C++, Unity, Node.js, Java, Go, Python SDKs, Rest and RPC APIs<br>

    you store data in documents that contain fields mapping to values<br>
    documents are stored in collections, which are containers for your documents<br>
    that you can use to organize your data and build queries<br>
    document supports many different data types, and you can create sub-collections within the documents<br>

    - Cloud firestore is serverless<br>
    theres no server to maintain, update, manage<br>
    - multi-region replication<br>
    - flexbility<br>
    - expressive querying<br>
    - realtime updates<br>
    - offline support<br>
    - Secure: firebase authentication, Cloud firestore security rules<br>
    - free tier - 1 db holding 5GB<br>
    - realtime database - simpler version of firestore<br>
<br>

**Firebase**<br>
firebase is a mobile application development platform that provide tools and cloud services<br>
that help enable developers to develop apps faster and easily<br>
<br>

- **Memorystore**<br>
    - Memorystore is a fully managed service for either Redis or Memcached in-memory data store to build apps caches<br>
    - fully managed<br>
    - replicated across zone, monitored for health, high availability<br>
    - scale as needed<br>
    - secure<br>
    - always up to date<br>
    - Use cases:<br>
        - caching<br>
        - gaming (leaderboards, user profiles)<br>
        - stream processing<br>
