### Cloud Storage Storage Types<br>

***

<br>

##### **Cloud Storage**<br>

- consistent, scalable, large-capacity, highly durable object storage (not file or block)<br>
objects = pics or vids, but not OS<br>
- Worldwide accessibility and worldwide storage locations<br>
- use for data files, text files, pictures, videos<br>
great for storing data for data analytics jobs, text files with codes etc..<br>
- Excels for content delivery, big data sets and backups<br>
- stored as objects in buckets<br>
<br>

**Cloud Storage Buckets**<br>

bucket basic container that holds your data<br>
everything you store in Cloud Storage must be contained in a bucket<br>
unlike folders and directory, you cannot nest buckets<br>

Upon creation: must have a global unique name<br>
geographic location: (Region, Dual-region, Multi-region)<br>
storage class: (Standard, Nearline, Coldline, Archive)<br>
access control: (Uniform(IAM), Fine-grained(IAM+ACLs))<br>
labels: optional<br>
<br>

**Cloud Storage objects**<br>

object(object data, metadata(properties: name, storage class etc))<br>
object -stores-in-> bucket<br>
<br>

##### **Storage Classes**<br>

**Standard (Hot data)**<br>
- Maximum availability<br>
- No storage duration<br>
- Analytical workloads and transcoding<br>
- $0.02 GB/mo<br>
- availability: >99.99% multi-regions, dual-regions, 99.99% regions<br>
- durability: 99.999999999% annual durability<br>

**Nearline (Hot data)**<br>
- Low-cost for infrequently accessed data<br>
- 30 days minimum storage duration<br>
- data backup and data archiving<br>
- $0.01 GB/mo<br>
continous file adding, but only access them once a month<br>
- availability: 99.95% multi-regions, dual-regions, 99.9% regions<br>
- durability: 99.999999999% annual durability<br>

**Coldline (Cold data)**<br>
- very low-cost for infrequent accessed data<br>
- 90 days minimum storage duration<br>
- data backup and data archiving<br>
- $0.004 GB/mo<br>
only access them once a quarter<br>
- availability: 99.95% multi-regions, dual-regions, 99.9% regions<br>
- durability: 99.999999999% annual durability<br>

**Archive (Cold data)**<br>
- lowest-cost archival storage<br>
- 365 days minimum storage duration<br>
- cold data storage disaster recovery<br>
- $0.0012 GB/mo<br>
only access them once a year<br>
- availability: 99.95% multi-regions, dual-regions, 99.9% regions<br>
- durability: 99.999999999% annual durability<br>
<br>

##### **Access Control**<br>

They can be used together<br>

**IAM**<br>
- Standard IAM permissions<br>
- permissions inherited hierarchically<br>

**Access Control List(ACL)**<br>
- defines who has access to your buckets<br>
and objects, as well as what level of access they have<br>

**Signed URLs**<br>
- time limited read/write access URL<br>
- access the object for the duration of time you specify<br>

**Signed Policy documents**<br>
- specify what can be uploaded to a bucket<br>

**IAM and ACLs**<br>
they work together<br>

IAM<br>
- only one has to grant the permission to access, but IAM is recommended<br>
- Two levels of granularity project or bucket level<br>
give perm to read in any bucket, but perm to create in 1 bucket<br>
- roles: primitive, standard, legacy. legacy are equivalent to ACLs<br>

ACL<br>
- granular permissions<br>
- each ACL entry = permission + scope<br>
- should be used with caution, as ACL overlap with IAM roles<br>

**Access Control**<br>

Signed URLs<br>
- access an object using the URL for the duration of time you specify<br>
- allow users without credentials to perform specific actions on a resource<br>
- actions are taken as a user or service account<br>
- do not need an account - just the URL<br>
> gsutil signurl -d 10m private-key.json gs://storage1/images/image.jpg

this will create a signed url for 10 minutes<br>
